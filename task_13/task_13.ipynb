# Series

It is like a single colunm in a tabular dataset

```python
import pandas as pd
student_marks = [71, 29, 73, 39, 36, 93, 51, 96, 86, 94, 94, 66, 6, 41, 30]
my_series = pd.Series(student_marks)
my_series
```

# Dataframe

It can be understood as combination of series

Creating a dataframe which will contain student marks and age

```python
# first we have to get data of student age (just to keep it simple I am generating it randomnly)
import numpy as np
student_ages = np.random.randint(18, 25, 15)
student_ages_saved = list(student_ages)
print(student_ages_saved)
```

```python
# creating series
student_marks_series = pd.Series(student_marks)
student_ages_series = pd.Series(student_ages_saved)

my_dataframe = pd.DataFrame({'Marks': student_marks_series,
                             'Age': student_ages_series})

my_dataframe
```
