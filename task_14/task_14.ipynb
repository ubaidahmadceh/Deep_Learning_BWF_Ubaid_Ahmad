# Reading sql

```python
import sqlite3
conn = sqlite3.connect('my_database.db')

data = {
  "Marks": [71, 29, 73, 39, 36, 93, 51, 96, 86, 94, 94, 66, 6, 41, 30],
  "Age": [19, 22, 23, 23, 24, 24, 23, 23, 18, 24, 19, 23, 20, 22, 22]
}

my_dataframe_2 = pd.DataFrame(data)

my_dataframe_2.to_sql('students', conn, if_exists='replace', index=False)

reading_sql = pd.read_sql_query('SELECT * FROM students', conn)
reading_sql
```
